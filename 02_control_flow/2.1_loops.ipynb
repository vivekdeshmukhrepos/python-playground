```python
## for loop
# This is similar to i==0 and i<5 so it will print from 0 to 4 
for i in range(5):
    print(i)
```

```python
# This is similar to i==1 and i<6 so it will print from 1 to 5 
for i in range(1,6):
    print(i)
```

```python
# This is similar to i==1. But increase i's value each time with +2
# start from 1 to 10
# So it will print odd numbers from 1,3,5,7 and 9

# range(start, stop, step)
for i in range(1,10,2):
    print(i)
```

```python
# This is similar to i==10. But increase i's value each time with -1
# So it will print odd numbers from 10,9,8,7,6,5,4,3,2
# Start from 10 to 1 
# range(start, stop, step)
for i in range(10,1,-1):
    print(i)
```

```python
# This is similar to i==10. But increase i's value each time with -2
# So it will print odd numbers from 10,8,6,4,2
# Start from 10 to 1 

for i in range(10,1,-2):
    print(i)
```

```python
# strings
# This will print each character in the string one by one
str="Vivek Deshmukh"

for i in str:
    print(i)
```

```python
## while loop

## The while loop continues to execute as long as the condition is True.

count=0

while count<5:
    print(count)
    count=count+1
```

```python
## Loop Control Statements

## break
## The break statement exits the loop permaturely

## break sstatement

for i in range(10):
    if i==5:
        break
    print(i)
   
```

```python
## continue

## The continue statement skips the current iteration and continues with the next.

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass
## The pass statement is a null operation; it does nothing.

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
## Nested loopss
## a loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## Examples- Calculate the sum of first N natural numbers using a while loop
 

counter=1
last_number=10
sum=0
while counter<=last_number:
    sum= sum+counter
    counter=counter+1
    
print("Sum of first 10 natural number:",sum)
```

```python
## Examples- Calculate the sum of first N natural numbers using a for loop

sum=0 
for i in range(11):
    sum=sum+i

print(sum)

# Using range(start, stop, step)

total = 0

for i in range(0, 11, 1):
    total += i

print(total)
```

```python
## Example- Prime numbers between 1 and 100

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else: # This belongs to for loop. It runs only if the for loop is not terminated by break
            print(num)
```
