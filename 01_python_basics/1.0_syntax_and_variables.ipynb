### Syntax in Python

```python
# Basic Syntax Rules In Python
# Case sensitivity- Python is case sensitive

# Declared two variables with names differing only in case
name="Vivek"
Name="John"

# Print both variables to show they are distinct
print(name)
print(Name)
```

```python
# This is sinle line comment
# If you want to write multiline comments in python you can use triple quotes As belows

'''
This is a example of multiline comments
You can keep on writing multiple lines
Line 1
Line 2 and so on
'''
# Print statement will not add \n in output
print(
'''
This is a example of multiline comments
You can keep on writing multiple lines
Line 1
Line 2 and so on
''')
```

```python
# Indentation is used to define blocks of code in Python

flag=True
if flag:
    print("This line is inside the if block")
    print("This line is also inside the if block")

print("This line is outside the if block")
```

```python
# Line Continuation
# Use a backslash (\) to continue a statement to the next line

total = 10 + 20 + 30 + \
        40 + 50

print(total)
```

```python
# Alternative: Using parentheses for line continuation
# Python also allows implicit line continuation inside brackets—no backslash needed:

total = (
    10 + 20 + 30 +
    40 + 50
)

print(total)
```

```python
# Multiple Statements on a single line
# Use a semicolon (;) to separate multiple statements on a single line

x = 5; y = 10; z = x + y
print(z)
```

```python
# Semantics in Python refers to the meaning and behavior of code, not just its syntax. 
# Even if the syntax is correct, incorrect semantics can lead to unexpected or wrong results.

# Example of semantic error
a = 10
b = "20"
# The following line has correct syntax but will raise a runtime error (semantic error)
# c = a + b  # This will cause a TypeError because you cannot add an integer and a string

# Correct Semantic Usage
c = a + int(b)  # Convert b to integer before addition
print(c)
```

```python
# Another Semantic Example

age = 18

# We should not use assignment (=) when we mean to compare
if age == 18:   # ✅ Correct semantic meaning. Using '==' for comparison
    print("Eligible")
```

```python
# Assigns the integer value 25 to the variable age.
age = 25

# Assigns the string "Alice" to the variable name.
name = "Alice"

# Returns the data type of the variable age.
type(age)
```

```python
# Returns the data type of the variable name.
type(name)
```

```python
## Dynamic Typing
## Python allows the type of a variable to change as the program executes
variable_name=10 #int
print(variable_name,type(variable_name))

variable_name="Hello" 
print(variable_name,type(variable_name))

variable_name=3.14
print(variable_name,type(variable_name))
```

```python
## Basic Calculator Example

# This program takes two numbers as input from the user and performs basic arithmetic operations such as addition, subtraction, multiplication, and division. 
# The results of each operation are then displayed on the screen.

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

sum_result = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("Sum:", sum_result)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)
```
