```python
### Accessing List Elements

fruits=["apple","banana","cherry","kiwi","gauva"]
print(fruits[0])
print(fruits[1])
print(fruits[2])
print(fruits[3])
print(fruits[4])
# print(fruits[5])    # This will give IndexError as there is no 6th element 
print(fruits[-1 ])  # Last element
print(fruits[-2 ])  # Second last element - If yo want to access elements from the end use - notation
```

```python
# Slicing Lists
print(fruits[1:]) # From index 1 to end
print(fruits[:3]) # From start to index 3-1
print(fruits[1:3]) # From index 1 to index 3-1
```

```python
# Modifying List Elements
print(fruits)
fruits[1]="watermelon"
print(fruits)
```

```python
# Modifying Multiple List Elements at Once 
# This replaces elements from index 1 onwards with the string "watermelon"
fruits[1:]="watermelon"
print(fruits)
print(fruits[2])
```

```python
# Reinitialize the list for next operations
fruits=["apple","banana","cherry","kiwi","gauva"]
fruits.append( "orange" )
fruits
```

```python
# Insert an element at specific position
# This will insert 'mango' at index 2 and shift the rest of the elements to the right
fruits.insert(2,"mango")
fruits
```

```python
## Remove and return the last element
popped_fruits=fruits.pop()
print(popped_fruits)
print(fruits)
```

```python
# Find the index of a specific element
index=fruits.index("cherry")
print(index)
```

```python
# insert another 'banana' at index 2 and count total 'banana' in the list
fruits.insert(2,"banana")
print(fruits)
print(fruits.count("banana"))

```

```python
# Remove all occurrence of 'banana' from the list
fruits = [w for w in fruits if w.lower() != "banana"]

print(fruits)
```

```python
# Sorts the list in ascending order
fruits.sort() 
fruits
```

```python
# REverse the list

fruits.reverse() 
fruits
```

```python
# Remove all items from the list

fruits.clear() 

print(fruits)
```

```python
# Slicing list
fruits=["apple","banana","cherry","kiwi","gauva"]
print(fruits[1:4])  # From index 1 to index 4-1
print(fruits[:3])   # From start to index 3-1
print(fruits[2:])   # From index 2 to end
print(fruits[-4:-1])# From index -4 to index -1-1
print(fruits[::2])  # Every second element
print(fruits[::-1]) # Reversed list
print(fruits[::-2]) # Reversed list with every second element
print(fruits[1:4:2])# From index 1 to index 4 -1 with step 2
```

```python
### Iterating Over List

for fruit in fruits:
    print(fruit)
```

```python
## Iterating with index
for index,fruit in enumerate(fruits):
    print(index,fruit)
```

```python
# List comprehension
# What is comprehension: It is a concise way to create lists in Python. 
# It consists of brackets containing an expression followed by a for clause, 
# and can also include optional if clauses to filter items.
# Create a list of squares of numbers from 0 to 9

lst=[]
for x in range(10): # Loop through numbers 0 to 9
    lst.append(x**2) # Calculate square and append to the list

print(lst)
```

```python
# List comprehension in short

[x**2 for x in range(10)]
```

```python
# List Comprehension with Condition
# Even numbers from 0 to 9
lst=[]
for i in range(10):
    if i%2==0:
        lst.append(i)

print(lst)
```

```python
# List Comprehension with Condition
# Even numbers from 0 to 9 
even_numbers=[num for num in range(10) if num%2==0]
print(even_numbers)
```

```python
## Nested List Comphrension

lst1=[1,2,3,4]
lst2=['a','b','c','d']

pair=[[i,j] for i in lst1 for j in lst2]

print(pair)
```

```python
## List Comprehension with function calls
words = ["hello", "world", "python", "list", "comprehension"]
lengths = [len(word) for word in words]
print(lengths)  # Output: [5, 5, 6, 4, 13]
```
